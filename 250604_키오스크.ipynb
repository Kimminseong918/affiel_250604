```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

        # 첫 주문
        n = 0
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                        print('HOT 음료를 주문')
                    elif t == 2:
                        self.temp = "ICE"
                        print('ICE 음료를 주문')
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
                print(self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 추가 주문 루프
        while n != 0:
            print()
            n = int(input('추가 주문은 음료 번호를, 지불은 0을 누르세요. : '))
            if n == 0:
                print('주문 완료')
                print(self.order_menu, self.order_price)
                print('총 합계 :', self.price_sum, '원')
                break #루프 탈출을 위해 추가됨
            elif 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                        print('HOT 음료를 주문')
                    elif t == 2:
                        self.temp = "ICE"
                        print('ICE 음료를 주문')
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
                print('추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계: ', self.price_sum, '원')
            else:
                print('없는 메뉴입니다. 다시 주문해주세요.')

    # 지불 메서드
    def pay(self):
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('결제 방식 선택: 현금(1번), 카드결제(2번): '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    # 주문서 출력
    def table(self):
        outline1 = '⟝' + '-' * 30 + '⟞'
        outline2 = '|' + ' ' * 31 + '|'

        print(outline1)
        for i in range(5):
            print(outline2)
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], " : ", self.order_price[i])
        print('합계 금액 : ', self.price_sum, '원')
        for i in range(5):
            print(outline2)
        print(outline1)
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

```python
### daytime 추가 ###



from datetime import datetime



class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0
        self.order_time = None

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

        # 첫 주문
        n = 0
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n <= len(self.menu):
                self.order_time = datetime.now()
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                        print('HOT 음료를 주문')
                    elif t == 2:
                        self.temp = "ICE"
                        print('ICE 음료를 주문')
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
                print(self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 추가 주문 루프
        while n != 0:
            print()
            n = int(input('추가 주문은 음료 번호를, 지불은 0을 누르세요. : '))
            if n == 0:
                print('주문 완료')
                print(self.order_menu, self.order_price)
                print('총 합계 :', self.price_sum, '원')
                break #루프 탈출을 위해 추가됨
            elif 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    self.order_time = datetime.now()
                    if t == 1:
                        self.temp = "HOT"
                        print('HOT 음료를 주문')
                    elif t == 2:
                        self.temp = "ICE"
                        print('ICE 음료를 주문')
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                self.price_sum += self.price[n-1]
                print('추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계: ', self.price_sum, '원')
            else:
                print('없는 메뉴입니다. 다시 주문해주세요.')

    # 지불 메서드
    def pay(self):
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('결제 방식 선택: 현금(1번), 카드결제(2번): '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    # 주문서 출력
    def table(self):
        outline1 = '⟝' + '-' * 30 + '⟞'
        outline2 = '|' + ' ' * 31 + '|'

        print(outline1)
        
        if self.order_time:
            print('| 주문일시 :', self.order_time.strftime('%Y-%m-%d %H:%M:%S').ljust(16), '|')
        else:
            print('| 주문일시 : -'.ljust(31), '|')
        for i in range(5):
            print(outline2)
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], " : ", self.order_price[i])
        print('합계 금액 : ', self.price_sum, '원')
        for i in range(5):
            print(outline2)
        print(outline1)
```
